```python
X=df.drop('total_bill', axis=1)
y=df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train ,y_test=train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
from sklearn.impute import SimpleImputer # for missing value but yha h ni man le agar hoti toj
from sklearn.preprocessing import OneHotEncoder # there are no mater the order that's why i am taking the label encodeing
from sklearn.preprocessing import StandardScaler

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
df.columns
```

```python
df
```

```python
cat_cols=['sex', 'smoker', 'day', 'time']
num_cols=['tip', 'size']

# don't scale the target variable
```

```python
num_pipeline=Pipeline(steps=[('imputation', SimpleImputer(strategy='median')),
                             ('scalling', StandardScaler())])

cat_pipeline=Pipeline(steps=[('imputation', SimpleImputer(strategy='most_frequent')),
                             ('encoding', OneHotEncoder())])
```

```python
preprocessor=ColumnTransformer([('num pipeline', num_pipeline, num_cols),
                   ('cat pipeline', cat_pipeline, cat_cols)])
```

```python
preprocessor
```

```python
X_train=preprocessor.fit_transform(X_train)
X_test=preprocessor.transform(X_test)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

models={'multiple linear Regression':LinearRegression(),
        'Supprot vector regression': SVR(),
        'DTR': DecisionTreeRegressor(),
        'Ramdom forest Regression': RandomForestRegressor()}
```

```python
from sklearn.metrics import r2_score
def model_train_eval(X_train,y_train, X_test, y_test, models): 
    evaluation={}
    for i in range(len(models)): 
        model=list(models.values())[i]
        model.fit(X_train, y_train)
        y_pred=model.predict(X_test)
        model_score=r2_score(y_test, y_pred)
        evaluation[list(models.keys())[i]]=model_score
    return evaluation
```

```python
model_train_eval(X_train, y_train, X_test, y_test, models)
```
