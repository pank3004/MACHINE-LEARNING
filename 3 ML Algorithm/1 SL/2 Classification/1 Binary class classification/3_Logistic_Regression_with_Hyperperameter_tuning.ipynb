```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
X,y=make_classification(n_samples=1000, n_features=10, n_redundant=5, n_informative=5,n_classes=2, random_state=1)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2, random_state=1)
```

```python
# hyperparameter tuning: 
     # 1 grid search cv
     # randomized search cv
```

```python
# 1 grid search cv

from sklearn.model_selection import GridSearchCV
```

```python
params={'penalty': ('l1', 'l2', 'elasticnet'), 'C': [1,2,10,20,30,40]}
# C means lembda
# 3*6=18 combination
```

```python
classifier=LogisticRegression()
```

```python
classifier
```

```python
clf=GridSearchCV(classifier, param_grid=params, cv=5)
# estimater=classifier 
# tatal model 3*5*5=75
```

```python
clf
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
model=LogisticRegression(C=1, penalty='l2')
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
model.predict_proba(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
# 2 randmozie search cv>> it will take randomly some combination (n_iter)

from sklearn.model_selection import RandomizedSearchCV
```

```python
randomized_clf=RandomizedSearchCV(classifier, param_distributions=params, cv=5, n_iter=10)
# 3*6=18 is the total combination of P and C in which it take 10 combination randomaly
```

```python
randomized_clf
```

```python
randomized_clf.fit(X_train, y_train)
```

```python
randomized_clf.best_params_
```

```python
randomized_clf.best_score_
```
