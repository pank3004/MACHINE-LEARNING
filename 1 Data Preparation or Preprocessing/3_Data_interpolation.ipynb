```python
#Data interpolation: is a process of estimating data bw the given range
```

```python
#1 Linear interpolation
#2 cubic 
#3 polynomial
```

```python
#Linear interpolation: 

import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)
```

```python
x_new=np.linspace(1,5,10)
```

```python
x_new
```

```python
y_interp=np.interp(x_new,x,y)
```

```python
y_interp
```

```python
plt.scatter(x_new, y_interp)
```

```python
# Cubic interpolation: 

x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
f=interp1d(x,y,kind='cubic')
```

```python
x_new=np.linspace(1,5,10)
```

```python
x_new
```

```python
y_interp=f(x_new)
```

```python
y_interp
```

```python
plt.scatter(x_new, y_interp)
```

```python
# polynomial interpolation

x=np.array([1,2,3,4,5])
y=np.array([1,4,9,1,25])
```

```python
p=np.polyfit(x,y,2)  # Degree of the fitting polynomial=2
                    # p is an array containing the coefficients of the polynomial that best fits the data points you provided
                    #Use np.polyfit to fit a polynomial of degree 2 to your data points. The function returns the coefficients of the polynomial.
```

```python
x_new=np.linspace(1,5,10)
```

```python
x_new
```

```python
y_interp=np.polyval(p,x_new) #Evaluate a polynomial at specific values
```

```python
y_interp
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new, y_interp)
```
