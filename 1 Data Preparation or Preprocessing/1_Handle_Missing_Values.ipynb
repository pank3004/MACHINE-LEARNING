```python
# whenever you have normal dist data, impute with mean
```

```python
df['mean_imputation']=df['age'].fillna(df['age'].mean())
```

```python
df[['age', 'mean_imputation']]
```

```python
sns.displot(df['mean_imputation'], kde=True)
```

```python
#mean imputation works well with normalyy distributed data or no outliear present
```

```python
# Median imputation
```

```python
df['median_imputation']=df['age'].fillna(df['age'].median())
```

```python
df[['age', 'median_imputation']]
```

```python
df[['age', 'mean_imputation', 'median_imputation']]
```
