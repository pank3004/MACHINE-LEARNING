```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
data=datasets.load_iris()
X=data.data
y=data.target
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# applying the pca: 
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_scaled=pca.fit_transform(X_scaled)
```

```python
pca_scaled
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1], c=y)
```

```python
# agglomerative clustering>> make dendogram: 

import scipy.cluster.hierarchy as sc

plt.figure(figsize=(20,7))
plt.title('dendrogram')

sc.dendrogram(sc.linkage(pca_scaled, method='ward'))
plt.xlabel('sample data')
plt.ylabel('Euclidean distance')
```

```python
# trick k=2
```

```python
from sklearn.cluster  import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
```

```python
cluster
```

```python
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1], c=cluster.labels_)
```

```python
# k starts from 2

from sklearn.metrics import silhouette_score

silhouette_coeff=[]
for k in range(2,11): 
    agglo=AgglomerativeClustering(n_clusters=k, metric='euclidean', linkage='ward')
    agglo.fit(pca_scaled)
    score=silhouette_score(pca_scaled, agglo.labels_)
    silhouette_coeff.append(score)
```

```python
silhouette_coeff
```

```python
plt.plot(range(2,11), silhouette_coeff)
plt.xticks(range(2,11))
plt.xlabel('number of clusters')
plt.ylabel('silhoutte coeff')
plt.show()
```

```python
# k=2 is optimal varified by silhouette score 
```
