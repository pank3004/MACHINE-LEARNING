```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
X,y=make_classification(n_samples=1000, n_features=10, n_redundant=5, n_informative=5,n_classes=2, random_state=1)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2, random_state=1)
```

```python
# hyperparameter tuning: 
     # 1 grid search cv
     # randomized search cv
```

```python
# 1 grid search cv

from sklearn.model_selection import GridSearchCV
```

```python
params={'penalty': ('l1', 'l2', 'elasticnet'), 'C': [1,2,10,20,30,40]}
# C means lembda
# 3*6=18 combination
```

```python
classifier=LogisticRegression()
```

```python
classifier
```

```python
clf=GridSearchCV(classifier, param_grid=params, cv=5)
# estimater=classifier 
# tatal model 3*6*5=90
```

```python
clf
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
clf.cv_results_['params']                                    
```

```python
clf.cv_results_['mean_test_score']


# By default, LogisticRegression() uses the lbfgs solver, which does not support:
            # 'l1' penalty
            # 'elasticnet' penalty
# So for these invalid combinations, GridSearchCV tries to fit the model, but it fails — and the result is NaN
```

```python
# mean_test_score: 

# This is the average score (like accuracy, precision, etc.) on the validation sets across all cv folds.
# It tells you how well the model performed, on average, with a particular set of hyperparameters.

# If cv=5, your model is trained 5 times on different splits, and the accuracy for a hyperparameter set was:
# [0.80, 0.82, 0.81, 0.79, 0.83]
# mean_test_score = (0.80 + 0.82 + 0.81 + 0.79 + 0.83) / 5 = 0.81
```

```python
clf.cv_results_['std_test_score']
```

```python
# std_test_score: 

# This is the standard deviation of the scores across all folds.
# It tells you how stable or consistent the model's performance is for a given parameter set.

# 📌 From the same example:
# std_test_score = standard deviation of [0.80, 0.82, 0.81, 0.79, 0.83]≈ 0.014
# A low std_test_score means the model performed consistently across folds — which is good.
# A high std_test_score means the model performance is not stable — maybe it overfits or is sensitive to the data.
```

```python
model=LogisticRegression(C=1, penalty='l2')
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
model.predict_proba(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
