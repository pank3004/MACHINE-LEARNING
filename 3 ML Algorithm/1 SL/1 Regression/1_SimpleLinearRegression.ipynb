Linear Regression models the relationship between the dependent variable (y) and one or more independent variables (x) by fitting a linear equation to observed data. In simple terms, it finds the best-fitting straight line through the data points

```python
data=pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
```

```python
data
```

```python
data['target']=diabetes.target
```

```python
data
```

```python
# 2 EDA & Datacleaning
# 3 data prepareation and feature engineering

# but hear already done
```

```python
#4 divide the data into X and y
# in simple linear regression 1 X and 1 y
X=data[['bmi']]
y=data[['target']]
```

```python
X
```

```python
y
```

```python
# 5 split the data into train and test

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
data.shape
```

```python
# 6 scalling the data: 
            # optional (already done in this case)
```

```python
# 7 Model Traning: 
from sklearn.linear_model import LinearRegression
```

```python
# instaniate an LR Model
model=LinearRegression()
```

```python
model
```

```python
# for training of model
model.fit(X_train, y_train)  
       # training of model>> try to understand the relationship bw X_train and y_train>> made Ordinary list square>>mean square error>> optimise by gradiant descent>> find optimal theta1 and theta0
       # at backend, it will use gradient descent to give optimaml coficient(theta1 and theta 0)
```

```python
model.coef_  # slop/m/B1/theta1 this is the optimal theta1
```

```python
model.intercept_  # c/B0/theta0 this is the optimal theta0
```

```python
# for prediction the data
y_pred=model.predict(X_test)  # will make the prediction by using theta1 and theta0 
```

```python
y_pred
```

```python
y_test
```

```python
# visualise the result: 
plt.scatter(X_test, y_test,color='black', label='Actual data')
plt.plot(X_test, y_pred, color='blue', linewidth=3, label='Linear regression line')
plt.xlabel('BMI')
plt.ylabel('One year progression-target')
plt.title('Linear Regression on Diabetes data')
plt.legend()
plt.show()
```

```python
#8 model evaluation
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)
```
