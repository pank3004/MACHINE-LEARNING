```python
# extreme values are outliers
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
salary = [11, 40, 45, 68, 65, 68, 78, 90, 57, 74, 91, 92, 88, 68, 57, 48, 99, 101, 68, 77, 110, 140]
```

```python
# five point summary
# min, 25th, 50th, 75th, 100th

np.quantile(salary, [0,0.25, 0.50, 0.75,1])
```

```python
df=pd.DataFrame(salary, columns=['salary'])
```

```python
df
```

```python
df.describe()
```

```python
# cheak outlirs>> 1. distplot, 2.boxplot
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df['salary'],kde=True)
plt.title("distribution plot")
plt.subplot(1,2,2)
sns.boxplot(data=df, x='salary')
plt.title("Box plot")
```

```python
### there are 3 technique to handle the outliers
#1 Dropping the outlier
#2 Capping the outlier
#3 Replace with mean or median
```

```python
# 1 Dropping the outlier

Q1=df['salary'].quantile(0.25)
Q3=df['salary'].quantile(0.75)

IQR=Q3-Q1

lower_fence=Q1-1.5*IQR
upper_fence=Q3+1.5*IQR
```

```python
lower_fence
```

```python
df_filtered=df[(df.salary>=lower_fence) & (df.salary<=upper_fence)]
```

```python
df_filtered
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df_filtered['salary'],kde=True)
plt.title("distribution plot")
plt.subplot(1,2,2)
sns.boxplot(data=df_filtered, x='salary')
plt.title("Box plot")
```

```python
# 2 Imputation with mean and median

df['salary_imputed_mean']=np.where((df.salary>=upper_fence) | (df.salary<=lower_fence), df['salary'].mean(), df['salary'])
```

```python
df
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df['salary_imputed_mean'],kde=True)
plt.title("distribution plot")
plt.subplot(1,2,2)
sns.boxplot(data=df, x='salary_imputed_mean')
plt.title("Box plot")
```

```python
df['salary_imputed_median']=np.where((df.salary>=upper_fence) | (df.salary<=lower_fence), df['salary'].median(), df['salary'])
```

```python
df
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df['salary_imputed_median'],kde=True)
plt.title("distribution plot")
plt.subplot(1,2,2)
sns.boxplot(data=df, x='salary_imputed_median')
plt.title("Box plot")
```

```python
# 3 Capping>> replacing with the nearest value which is not outlier
```

```python
df
```

```python
lower_cap=df['salary'].quantile(0.05) # lower cap as 5th percentile (let)
upper_cap=df['salary'].quantile(0.95)  # upper cap at 95th percentile
```

```python
lower_cap
```

```python
upper_cap
```

```python
# replace outliers with the cap

df['salary_capped']=np.where(df['salary']<lower_cap, lower_cap, np.where(df['salary']>upper_cap, upper_cap, df.salary))
```

```python
df
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df['salary_capped'],kde=True)
plt.title("distribution plot")

plt.subplot(1,2,2)
sns.boxplot(data=df, x='salary_capped')
plt.title("Box plot")
```
