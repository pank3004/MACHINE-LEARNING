```python
data=load_iris()
```

```python
print(data.DESCR)
```

```python
X=df
y=data.target
```

```python
X
```

```python
y
```

```python
# train test split: 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#criterion{“gini”, “entropy”, “log_loss”}, default=”gini”
classifier=DecisionTreeClassifier(criterion='entropy')
```

```python
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier, filled=True)

plt.show()
```

```python
np.unique(X_train['petal width (cm)'])
```

```python
import numpy as np


# Step 1: Sort unique values
unique_values = np.unique(X_train['petal width (cm)'])
sorted_unique_values = np.sort(unique_values)

# Step 2: Calculate midpoints
split_points = []
for i in range(len(sorted_unique_values) - 1):
    midpoint = (sorted_unique_values[i] + sorted_unique_values[i + 1]) / 2
    split_points.append(midpoint)

# Print the possible split points
print("Possible split points:", split_points)
```

```python
# we treated each value of the possible split points(sort unique value>> calculate mid point bw each pair of consecutive uniquc values)
#  as a cutoff/threshold and try to build a decision tree and whichever threshold gives you the best information gain that
#  will be use for splitting
```

# postpruning: 


```python
# postpruning: 
classifier=DecisionTreeClassifier(criterion='entropy', max_depth=3)
```

```python
classifier.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier, filled=True)

plt.show()
```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# prepruning using hyperparameter tuning


```python
#prepruning using hyperparameter tuning
parameter={'criterion': ['gini', 'entropy', 'log_loss'],
        'splitter': ['best', 'random'],
        'max_depth': [1,2,3,4,5,6],
        'max_features': ['sqrt', 'log2', 'auto']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf=DecisionTreeClassifier()
```

```python
model=GridSearchCV(clf, param_grid=parameter, cv=5, scoring='accuracy')
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```
