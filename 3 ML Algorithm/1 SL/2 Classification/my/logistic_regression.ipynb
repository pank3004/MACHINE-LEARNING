```python
# Assumption

# I assume that the data are missing completely at random (MCAR). 
# There are two methods which can be used to impute missing values. 
# One is mean or median imputation and other one is random sample imputation. 
# When there are outliers in the dataset, we should use median imputation. 
# So, I will use median imputation because median imputation is robust to outliers.

# I will impute missing values with the appropriate statistical measures of the data, in this case median.
#  mputation should be done over the training set, and then propagated to the test set. 
# It means that the statistical measures to be used to fill missing values both in train and test set, should be extracted from the train set only.
#  This is to avoid overfitting
```

```python
# for col in numericals:
#     med=X_train[col].median()
#     X_train[col].fillna(med, inplace=True)
#     X_test[col].fillna(med, inplace=True)

for col in numericals:
    med=X_train[col].median()
    X_train.fillna({col:med}, inplace=True)
    X_test.fillna({col:med}, inplace=True)

```

```python
# Engineering outliers in numerical variables: 

# We have seen that the Rainfall, Evaporation, WindSpeed9am and WindSpeed3pm columns contain outliers. 
# I will use top-coding approach to cap maximum values and remove outliers from the above variables.
```

```python
df.columns
```

```python
outlies < -1.2000000000000002 &&   outliers > 2.0 >>> Rainfall
outlies < -4.600000000000001 &&   outliers > 14.600000000000001>>> Evaporation
outlies < -11.0 &&   outliers > 37.0  >>> windspeed9am
outlies < -3.5 &&   outliers > 40.5 >> windspeed3pm

```

```python
def max_value(df3, col, top): 
    return np.where(df3[col]>top, top, df3[col])

for df3 in [X_train, X_test]: 
    df3['Rainfall']=max_value(df3, 'Rainfall', 2.0)
    df3['Evaporation']=max_value(df3, 'Evaporation', 14.60)
    df3['WindSpeed9am']=max_value(df3, 'WindSpeed9am', 37.0)
    df3['WindSpeed3pm']=max_value(df3, 'WindSpeed3pm', 40.5)

```
