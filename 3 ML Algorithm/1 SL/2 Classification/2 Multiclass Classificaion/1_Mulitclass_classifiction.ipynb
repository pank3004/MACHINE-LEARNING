```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import make_classification
X,y=make_classification(n_samples=1000, n_features=10,n_informative=5, n_redundant=5,n_classes=3, random_state=1)
```

```python
y
```

```python
X
```

```python
# multiclassification problem: 

# 1 one vs test
# 2 multinomial method softmax regression
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.30, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#mult_class>>  ovr, multinomial, auto by default this will be auto
# solver{‘lbfgs’, ‘liblinear’, ‘newton-cg’, ‘newton-cholesky’, ‘sag’, ‘saga’}, default=’lbfgs’>Algorithm to use in the optimization problem.
# solver is help to find the optimal coffcients(thetas)
model=LogisticRegression(multi_class='ovr',solver='lbfgs')
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
y_pred_proba=model.predict_proba(X_test)
```

```python
y_pred_proba
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
