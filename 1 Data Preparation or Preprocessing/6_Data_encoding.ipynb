Ordinal encoding

```python
# (ii) ordinal encoding: 

# high school: 1
# graduate: 2
# post gra. 3
# phd: 4

# there are hierarchy
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({'qualification': ['HS', 'GR', 'PG', 'PHD','HS', 'GR','PG','PHD','PG', 'PG', 'PHD']})
```

```python
df
```

```python
ordinal_encoder=OrdinalEncoder(categories=[['HS', 'GR', 'PG', 'PHD']])
                                        #   [0,    1,    2,    3   ]
```

```python
ordinal_encoder.fit_transform(df[['qualification']])
```

```python
#for new
ordinal_encoder.transform([['PHD']]) 
```

target guided ordinal encoding

```python
# 3 target guided ordinal encoding


# relationship with target variable
# a lot of uniqe categoris
# replace the cagtegory with the mean and median of respective group target
```

```python
df=pd.DataFrame({'time':['lunch','breakfast', 'dinner','lunch','lunch','breakfast', 'dinner','lunch', 'lunch','breakfast', 'dinner'],
              'total_bill': [120,130,90,125,150,190,160,180,189,124,165]})
```

```python
df
```

```python
df.groupby('time')
```

```python
df.groupby('time')['total_bill']
```

```python
df.groupby('time')['total_bill'].mean()
```

```python
mean_price=df.groupby('time')['total_bill'].mean().to_dict()
```

```python
mean_price
```

```python
df['time_encoded']=df['time'].map(mean_price)
```

```python
df
```
