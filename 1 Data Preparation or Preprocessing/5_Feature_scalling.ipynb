```python
import seaborn as sns
import pandas as pd
import numpy as np
# import warnings
# warnings.filterwarnings('ignore')
```

```python
#feature scalling: 

# 1. standardisation >> mu=0, sigma=1
# 2. Normalization(minmax scalling)>> data come bw[0,1]
# 3. unit vector
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
mean=np.mean(df['total_bill'])
```

```python
std=np.std(df['total_bill'])
```

```python
mean
```

```python
std
```

```python
std_data=[]
for i in list(df['total_bill']): 
    zscore=(i-mean)/std
    std_data.append(zscore)
```

```python
print(std_data)
```

```python
np.mean(std_data)      # almost 0
```

```python
np.std(std_data)
```

```python
# featured scalling is option>> because it doesn't change the distribution of data
```

```python
sns.histplot(df['total_bill'], kde=True)
```

```python
sns.histplot(std_data, kde=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill']])  # it will calculate mu and sigma
                                # the use of double square brackets is a way to select a DataFrame rather than a Series in pandas
```

```python
scaler.transform(df[['total_bill']])  # apply the calculation on the dataset using mu and sigma
```

```python
df
```

```python
scaler.fit_transform(df[['total_bill','tip']]) # both fit and tranform in one line
```

```python
pd.DataFrame(scaler.fit_transform(df[['total_bill','tip']]), columns=['total_bill', 'tip'])
```

```python
df1=pd.DataFrame(scaler.fit_transform(df[['total_bill','tip']]), columns=['total_bill', 'tip'])
```

```python
df1['total_bill'].mean()
```

```python
df1['total_bill'].std()
```

```python
scaler.transform([[13,1]])  # 13 total bill and 1 is the tip
                            # only Transform because this is unseen data/test data
```
